```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv(r'income.data.csv')
df
```

```python
plt.scatter(df.income, df.happiness)
plt.title("The Correlation between Income and Happiness")
plt.ylabel("Happiness")
plt.xlabel("Income")
plt.show()
```

```python
data = df.iloc[:,1:-1]
label = df.iloc[:,-1]


x_train , x_test , y_train , y_test = train_test_split(data,label , test_size=0.5 )
```

```python
reg = LinearRegression()

reg.fit(x_train, y_train)

print (reg.coef_)
print (reg.intercept_)

print (f"y={reg.coef_[0]:.2f}x + {reg.intercept_:.2f}")


pre =reg.predict(x_test)

df2 = pd.DataFrame({"real":y_test , "pre":pre})
print (df2)

print (mean_absolute_error(y_test,pre))
print (mean_squared_error(y_test, pre))
```

```python
import numpy as np

plt.scatter(df.income, df.happiness)

x = np.arange(2,7,0.0000005)
y = reg.coef_ * x  + reg.intercept_
plt.plot(x,y , "r")
```
